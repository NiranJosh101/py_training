# Conditional Statement

```python
# if statement
age=20

if age >= 18:
    print("You are allow to vote")
```

```python
## else statement
age=16

if age >= 18:
    print("You are allow to vote")
else:
    print("Not old enough")
```

```python
#elif statement
age=15

if age<13:
    print("You are a child")
elif age%2 != 0: # all to check multiple conditions
    print("You are a teenager")
else:
    print("You are an adult")
```

```python
## Nested conditional statements
# tell if number is even, odd, or nagative

num=int(input("Enter the number"))

if num>0:
    print("The number is positive")
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("The number is negative")

```

```python
## Practical Example

## Determine if a year is a leap year i.e year/4 remaining 0

year = int(input("Enter the year here"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print("Year is a leap year")
        else:
            print("Year not a leap year")
    else:
        print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```
