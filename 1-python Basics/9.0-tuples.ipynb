# Tuples

```python
## creating a tuple
em_tuple=()
print(em_tuple)

tpl=tuple()
print(tpl)
```

```python
numbers=tuple([1,2,3,4,5,6]) # using a list
numbers
```

```python
mixed_tul=(1,"Hello world",3.14,True)
print(mixed_tul)
```

```python
## Accessing tuple elements
numbers=tuple([1,2,3,4,5,6])
print(numbers[3])
print(numbers[-1])
```

```python
numbers[0:3]
```

```python
## Tuple Operations

concat_tuple=numbers + mixed_tul
print(concat_tuple)
```

```python
mixed_tul * 3
```

```python
## immutable nature of tuples

lst=[1,2,3,4,5]
print(lst)

lst[2] = "Josh"
print(lst)
```

```python
numbers=tuple([1,2,3,4,5,6])
print(numbers)

# numbers[0] = "Josh" # can assign to a tuple element
```

```python
## Tuple methods
numbers=tuple([1,1,2,7,4,5,6])
print(numbers.count(1)) # count the number of times an element is present
print(numbers.index(7)) # check location of a particular element
```

```python
## packing and unpacking tuples

packed_tul= 1, "hello", 3.14
print(packed_tul)

```

```python
## unpacking a tuple

packed_tul= (1, "hello", 3.14)
print(packed_tul)


a,b,c=packed_tul
print(a)
print(b)
print(c)
```

```python
## unpacking with *
numbers=(1,2,3,4,5,6)
first, *middle, last = numbers

print(first)
print(middle)
print(last)
```

```python
## Nested tuple

# nested list
lst=[[1,2,3,4],[6,7,8,9],[1, "hello", 3.14, "c"]]
lst[2][1]
```

```python
## Nested tuple
lst=[[1,2,3,4],[6,7,8,9],(1, "hello", 3.14, "c")]
print(lst[2][0:2])

nested_tuple = ((1,2,3), ("a","b","c"), (True, False))
print(nested_tuple[1])
print(nested_tuple[1][2])

```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```
