# Decorators
#### They are used to modify the bahavior of a function or class method.

```python
## function copy
def welcome_test():
    return "Welcome to the advance python course"

welcome_test()
```

```python
# copying one function to another
wel=welcome_test
print(wel())

# delete initial function
del welcome_test
print(wel())
```

```python
# Closures

def main_welcome(msg):
    # msg='Welcome'

    # closure function
    def sub_welcome_method():
        nonlocal msg
        print('Welcome to the advance python course')
        print(msg)
        print('Please learn this concept properly')
    return sub_welcome_method()
```

```python
main_welcome('Welcome everyone')
```

```python
def main_welcome(func):
    # msg='Welcome'

    # closure function
    def sub_welcome_method():
        print('Welcome to the advance python course')
        func('welcome guys')
        print('Please learn this concept properly')
    return sub_welcome_method()
```

```python
main_welcome(print)
```

```python
def main_welcome(func, list):
    # msg='Welcome'

    # closure function
    def sub_welcome_method():
        print('Welcome to the advance python course')
        print(func(list))
        print('Please learn this concept properly')
    return sub_welcome_method()
```

```python
main_welcome(len,[1,2,3,4,5])
```

```python
### Decorator
def main_welcome(func):
    # msg='Welcome'

    # closure function
    def sub_welcome_method():
        print('Welcome to the advance python course')
        func()
        print('Please learn this concept properly')
    return sub_welcome_method()
```

```python
def course_intro():
    print("This is an advance py course")


course_intro()
```

```python
main_welcome(course_intro)
```

```python
## to create a decorator so I can display messages without call the main function
@main_welcome
def course_intro():
    print('This is an advance py course')
```

```python
## practical example for decorator
def my_deco(func):
    def wrapper():
        print('Something is haappening before the function is called')
        func()
        print('Somethingis happening')
    return wrapper
```

```python
@my_deco
def say_hello():
    print('Hello!')
```

```python
say_hello()
```

```python
## Decorator with argument
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print('Hello')
```

```python
say_hello()
```
