# Generators

#### They are a simpler way to create iterators. They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory

```python
# using generator we have created iterators
def square(n):
    for i in range(3):
        yield i**2
```

```python
square(3)
```

```python
# to call the generator
for i in square(3):
    print(i)
```

```python
# another way to call the generator
a=square(3)
a
```

```python
next(a)
```

```python
def my_gene():
    yield 1
    yield 2
    yield 3
```

```python
gen=my_gene()
gen
```

```python
next(gen)
```

```python
# testing with for loop
for val in gen:
    print(val)
```

```python
### Practical Example: Reading Large Files

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
file_path='large_file.txt'

# calling the generator
for line in read_large_file(file_path):
    print(line.strip())
```
