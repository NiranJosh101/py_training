# Abtstraction

```python
# is the concept of hiding the complex implementation details and showing only the necessary feature of an object. This helps in reducing programming complexity and effort
```

```python
from abc import ABC,abstractmethod
```

```python
## Abstract class
class Vehicle(ABC):
    def drive(self):
        print('The vehicle is used for driving')

    # an empty shell is created in the base class
    @abstractmethod
    def start_engine(self):
        pass


class Car(Vehicle):

    # the empty can be filled however by the derive class. That's abstraction
    def start_engine(self):
        print('Car engine started')


# def operate_vehicle(vehicle):
#     vehicle.start_engine()

car=Car()
car.start_engine()
# operate_vehicle(car)
```
