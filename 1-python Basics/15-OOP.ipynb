```python
## Instance Variable and Methods
class Dog:
    ## constructor just like a parameter
    def __init__(self,name,age):

        # creating the instance variable
        self.name=name
        self.age=age

## Create objects
dog1=Dog('Josh', 25)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
dog2=Dog('Lucy',4)
print(dog2)
print(dog2.name)
print(dog2.age)
```

```python
## Define a class with instance method
class Dog:
    ## constructor just like a parameter
    def __init__(self,name,age):

        # creating the instance variable
        self.name=name
        self.age=age

    # create instance method which are like added properties
    def bark(self):
        print(f"{self.name} says woof!")

dog3=Dog('Jerry', 1)
dog3.bark()

dog4=Dog('Jack', 5)
dog4.bark()
```

```python
### Modeling a Bank Acct


## Define a class for bank acct
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f'{amount} is deposited. New balance is {self.balance}') # using self.balance bc balance is a instance variable from somewhere else

    def withdraw(self,amount):
        if amount>self.balance:
            print("Insuffcient funds!")
        else:
            self.balance-=amount
            print(f'{amount} is withdrawn. New Balance is {self.balance}')

    def get_balance(self):
          return self.balance
    

## create an account

account=BankAccount('Josh', 5000)
print(account.balance)
```

```python
## Call instance method
account.deposit(100)
account.withdraw(200)
account.get_balance()
```
