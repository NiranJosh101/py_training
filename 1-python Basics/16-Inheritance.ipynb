# Inheritance

```python
## Inheritance
## Define parent class
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype

    def drive(self):
        print(f'The person will drive the {self.enginetype} car')
```

```python
car1=Car(4,5,'petrol')
car1.drive()
```

```python
## defining inheritance (Single Inheritance)
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):

        #calling the parent class from Car class
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving=is_selfdriving
    
    def selfdriving(self):
        print(f'Tesla support self driving : {self.is_selfdriving}')
```

```python
## Calling
tesla1=Tesla(4,5,'electric',True)
tesla1.selfdriving()
tesla1.drive()
```

```python
## defining inheritance (Multiple Inheritance)
## When a class inherits from more than one base class

# Base class 1
class Animal:
    def __init__(self,name):
        self.name=name

    def specker(self):
        print('Subclass must implement this methos')
    
## Base class 2
class Pet:
    def __init__(self,owner):
        self.owner=owner


# Derived class (inherits from multiple base classes)
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name) # you can call the class direct instead of super().__init__ method
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} says WOOF!"


# Create object

doggy=Dog('Jerry','Josh')
print(doggy.speak())
print(doggy.specker())
print(f'Owner : {doggy.owner}')
```
