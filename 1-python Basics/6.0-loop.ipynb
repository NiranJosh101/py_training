# Loops in python

```python
## for loop

for i in range(5):
    print(i)
```

```python
for x in range(20):
    print(x)
```

```python
x=1
y=101



for i in range(x,y):
    print(i)
```

```python
for i in range(1,10,2): # the last figure tell how the iteration should happen
    print(i)
```

```python
for i in range(2, 8, 3):
    print(i)
```

```python
## strings

name="Josh Niran"
for i in name:
    print(i)
```

```python
## while loop

## continues to execute as long as the condition is True.

count = 0

while count<5:
    print(count)
    count+=1
```

```python
count=0
while count%2==0:
    print(count)
    count+=1
```

```python
## Loop control statement

## break: The break statement exits the loop prematurely

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
x=2
y=3

for i in range(10):
    if i > x*y:
        break
    print(i)
```

```python
## continue

## The continue statement skip the current iteration and continues
   # with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
    
```

```python
x=2
y=1

for i in range (10, 20):
    if i%x*y==0:
        continue
    print(i)
```

```python
## pass
## The pass statement is a null operation; it does nothing

for i in range(5):
    if i==3:
        print("the number is", i)
        pass
    print(i)
```

```python
## Nested loops
## a loop inside a loop

for i in range(3):
    for j in range(2):
        for z in range(1):
            print(f"i:{i} and j:{j} and z:{z}")
```

```python
## Example- Calculate the sum of first N natural numbers 
#  using a while and for loop

## while loop

n=10
sum=0
count=1

while count<=n:
    sum=sum+count
    count+=1
print(f"Sum of first 10 natural number: {sum}")
```

```python
x = 20
y= 0
z = 3

while z<x:
    y += z
    z+=1
print(y)
```

```python
sum=0

for i in range(11):
    sum+=i
print(sum)
```

```python
## Example- prime number btw 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```

```python
for x in range(1, 201):
    if x>1:
        for i in range(2,x):
            if x%i==0:
                break
        else: print("Prime:", x)
```
