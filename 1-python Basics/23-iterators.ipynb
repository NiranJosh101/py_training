# Iterators
#### They are advance python concepts that allow for efficient looping and memory management. Iterators provide a way to access element of a collection sequentially without exposing the underlying structure

```python
# regular looping
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)

type(my_list)
print(my_list)
```

```python
## Iterator
iterator=iter(my_list)
# print(type(iterator))

iterator
```

```python
## to iterate through all the element
# next(iterator)
```

```python
iterator=iter(my_list)
```

```python

try:
    print(next(iterator))
except StopIteration:
    print('There are no more elements in the iterator')
```

```python
# str iterator
my_string = "Hello"
string_iterator = iter(my_string)

try:
    print(next(string_iterator))
except StopIteration:
    print('There are no more elements in the iterator')

print(next(string_iterator)) # Output: H
print(next(string_iterator)) # Output: e
```
