# Exception Handling 

#### Exception are events that disturp the normal flow of a program. They occur when an error is encountered during program execution.

```python
## Exceeption try, except block

try:
    a=b
except:
    print("The variable has not been assigned")
```

```python
try:
    a=b
except NameError as ex: # the error is caught here
    print(ex)
```

```python
try:
    result = 1/0
except ZeroDivisionError as zd:
    print(zd)
    print('Please enter the rigth divider')
```

```python
try:
    result = 1/2
    a=b
except ZeroDivisionError as zd:
    print(zd)
    print('Please enter the rigth divider')
except Exception as exa: # handles every possible error
    print (exa)
```

```python
try:
    num=int(input('Enter a number'))
    result=10/num
    print(result)
except ValueError:
    print('This is not a valid number')
except ZeroDivisionError:
    print('Enter a denominator > 0')
except Exception as ex: # to catch unexpected errors
    print(ex)
```

```python
## try,except,else block

try:
    num=int(input('Enter a number:'))
    result = 10/num
except ValueError:
    print('This is not a valid number')
except ZeroDivisionError:
    print('Enter a denominator > 0')
except Exception as ex: # to catch unexpected errors
    print(ex)
else:
    print(f'the result is {result}')
```

```python
## try,except,else and finally

try:
    num=int(input('Enter a number:'))
    result = 10/num
except ValueError:
    print('This is not a valid number')
except ZeroDivisionError:
    print('Enter a denominator > 0')
except Exception as ex: # to catch unexpected errors
    print(ex)
else:
    print(f'the result is {result}')
finally:
    print('Execution complete')
```

```python
### File handling and Exception handling

try:
    file=open('example.txt','r')
    content=file.read()
    print(content)
except FileNotFoundError:
    print('The file does not exist')
finally:
    if 'file' in locals() or not file.closed():
        file.close()
        print('file close')
```
