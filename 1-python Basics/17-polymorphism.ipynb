# Polymorphism

```python
## Method Overriding
## This allows a child class to provide a specific implementation of a method that is already defined in its parent class.

## Base class
class Animal:
    def speck(self):
        return "Sound of the animal"


## Derived class 1
class Dog(Animal):
    def speak(self): # the method alread exist but the child class redefines it
        return 'WOOF!'


## Derived class
class Cat(Animal):
    def speak(self):
        return 'Meow!'


dog=Dog()
cat=Cat()
print(dog.speak())
print(cat.speak())


```

```python
## Function that demostrates polymorphism
def animal_speak(animal):
    print(animal.speak())

animal_speak(dog)
```

```python
## Poly. with Functions and Methods
## base class

class Shape:
    def area(self):
      return 'The area of the figure'  

## Derive class 1
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height

    def area(self):
        return self.width * self.height


## Derive class 2
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14*self.radius *self.radius



## Functions that demostrate poly.
def print_area(shape):
    print(f'The Area is {shape.area()}')


rectangle=Rectangle(4,5)
circle=Circle(36)

print_area(rectangle)
print_area(circle)
```

```python
## Poly. with Abstract Base Classes
from abc import ABC,abstractmethod

## Define an abstarct class (they are like empty classes)
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass


## Derived class 1
class Car(Vehicle):
    def start_engine(self):
        return 'Car engine started'


## Derive class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return 'Motorcycle engine started'


## Function that demostrates poly.
def start_vehicle(vehicle):
    print(vehicle.start_engine())


## create objects
car=Car()
moto=Motorcycle()

start_vehicle(car)
start_vehicle(moto)

```
