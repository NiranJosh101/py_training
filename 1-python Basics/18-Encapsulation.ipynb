# Encapsulation and Abstation

```python
## Encapsulation is the concept of wrapping data (variables) and the method (functions) together as a single unit.
## It restricts direct access to some of the object's components

## Encapsulation with Getter and Setter Methods
## Public, Protected, Private variabless


class Person:
    def __init__(self,name,age):
        self.name=name ## public variable
        self.age=age ## public varaible


def get_name(person):
    return person.name


person=Person('Josh',25)
get_name(person)
```

```python
class Person:
    def __init__(self,name,age,gender):
        self.__name=name ## private variable
        self.__age=age ## private variable
        self.gender=gender ## public varible


def get_name(person):
    return person.__name


person=Person('Josh',25, 'Male')
# get_name(person)
```

```python
## Protected variables
# the variable can't be accessed outside the class but can via the derive class
class Person:
    def __init__(self,name,age,gender):
        self._name=name ## protected variable
        self._age=age ## protected variable
        self.gender=gender ## public varible

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)



employee=Employee('Dave',25,'Female')
print(employee._name)
```

```python
class Person:
    def __init__(self, name, age):
        self.__name = name  # private variable
        self.__age = age    # private variable

    # Getter for name
    def get_name(self):
        return self.__name

    # Setter for name
    def set_name(self, name):
        self.__name = name

    # Getter for age
    def get_age(self):
        return self.__age

    # Setter for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print('Age cannot be negative.')


# Usage
person = Person('Ifaka', 34)

print(person.get_name())   # Output: Ifaka
print(person.get_age())    # Output: 34

person.set_age(40)
print(person.get_age())    # Output: 40

person.set_age(-5)         # Output: Age cannot be negative.

person.set_name('Josh')
print(person.get_name())
```
