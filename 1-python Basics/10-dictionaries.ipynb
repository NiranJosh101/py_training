# Dictionaries

```python
## creating Dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
# key-value pair
student={"name":"Josh","age":12,"grade":24} # key are unchangeable but values are
print(student)
```

```python
## accessing dict elements
student={"name":"Josh","age":12,"grade":"A"}
print(student)

print(student["grade"])
print(student["name"])
```

```python
## Accessing using get() method
print(student.get("grade"))
print(student.get("last_name"))
print(student.get("last_name","Not Available")) # add a defualt value
```

```python
## Modifying dict element
## Dict are mutable, you can add, update and delete

print(student)
```

```python
## updated the value of dict

student["age"]=33
print(student)
student["address"]="Nigeria" # added new value and key
print(student)
```

```python
## delete dict element
del student["grade"] ## delete key value pair
print(student)
```

```python
## Dict methods

keys=student.keys() # get all key
print(keys)

values=student.values() # get all value
print(values)

items=student.items() #get all key value pair
print(items)
```

```python
## shallow copy
student_copy=student
print(student_copy)
```

```python
student["name"]="Krish"
print(student)
print(student_copy)
```

```python
## shallow copy
student_copy1=student.copy() # allow proper copying
print(student_copy1)
print(student)
```

```python
student["name"]="Dave"
print(student_copy1)
print(student)
```

```python
## iterating over dict
# you can use loops to iterate over dict, keys, values or items

# iterating over keys
for keys in student.keys():
    print(keys)
```

```python
# iterate over values
for value in student.values():
    print(value)
```

```python
# iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested dict
students={
    "student1":{"name":"Dave","age":32},
    "student2":{"name":"Josh","age":25}
}
print(students)
```

```python
## access nested dict elements
print(students["student2"]["name"])
print(students["student1"]["age"])
```

```python
## iterated over nested dict

for student_id,student_info in students.items():
    print(f"student{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## dict comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## conditional dict conprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Example

## use a dict to count the frequency of element in list

numbers=[1,2,2,3,3,3,4,4,4,4]
freq={}

for number in numbers:
    if number in freq:
        freq[number]+=1
    else:
        freq[number]=1
print(freq)
```

```python
## merge 2 dict into one

dict1={"a":1, "b":2}
dict2={"b":3, "c":4}

merge_dict={**dict1,**dict2}
print(merge_dict)
```
