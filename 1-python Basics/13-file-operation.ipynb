# File Operation

```python
### Read a whole file

with open('source.txt','r') as file:
    content=file.read() # the entire content would be read
    print(content)
```

```python
with open( 'source.txt','r') as file:
    content=file.readline()# the first line would be read
    print(content)
```

```python
## Read a file line by line
with open('source.txt','r') as file:
    for line in file:
        print(line.strip()) # strip() removes newline character
```

```python
## Writing a file(also overwriting)

with open('source.txt','w') as file:
    file.write('It will be great!\n')
    file.write('This is a new line')
```

```python
## Write a file(without overwriting)
with open('source.txt','a') as file:
    file.write('Append operation taking place!\n')
```

```python
## Writing a list of lines to a file
lines=['First line \n', 'Second line \n', 'Third line \n']

with open('source.txt', 'a') as file:
    file.writelines(lines)
```

```python
## Binary files

# Writing to a binary file
data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file: # 'wb' is write byte
    file.write(data)
```

```python
## Reading a binary file

with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
## Example
## Read the content from one  file and write it in a destination text file

# Copying a text file
with open('source.txt', 'r') as source_file:
    content = source_file.read()

# writing in destination file
with open('destination.txt', 'w') as destinaation_file:
    destinaation_file.write(content)
```

```python
## Writing and then reading a file

with open('source.txt','w+')  as file:
    # w+ used for both reading and writing
    file.write('Hello World \n')
    file.write('This is a new line \n')

    ## Move the file cursor to the beginning
    file.seek(0)

    # Read the content of the file
    content=file.read()
    print(content)
```

# Working with file paths

```python
import os

## Create a new directory
new_directory='File_path'
os.mkdir(new_directory)
print(f"Directory '{new_directory}' create")
```

```python
## Listing files and directories
items=os.listdir('.')
print(items)
```

```python
## Joining Paths

# adding file to a folder
dir_name='folder'
file_name='file.txt'
full_path=os.path.join(dir_name,file_name)
print(full_path)
```

```python
# adding file to a folder and get the folder location
dir_name='folder'
file_name='file.txt'
full_path=os.path.join(os.getcwd(),dir_name,file_name)
print(full_path)
```

```python
# check to see if a path exsit or not
path='source.txt'
if os.path.exists(path):
    print(f"The path '{path}' exists")
else:
    print(f"The path '{path}' does not exist")
```

```python
## Getting the absolute path
relative_part='example.txt'
absolute_path=os.path.abspath(relative_part)
print(absolute_path)
```
