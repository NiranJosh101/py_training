# Sets

```python
## Create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
# using set constructor
m_set=set([1,2,3,3,4,5,6]) # no duplicate elements allowed
print(m_set)
```

```python
## basic set operations

## Adding and removing elements
m_set.add(7)
print(m_set)
```

```python
## Removing the element from a set

m_set.remove(3)
print(m_set)
```

```python
## not return error if you remove element not present

my_set.discard(11)
print(my_set)
```

```python
## pop method
remove_element=m_set.pop() # remove first element
print(remove_element)
print(m_set)
```

```python
## clear all elements
my_set.clear()
print(my_set)
```

```python
## Set membership test
## check if an element is a member of a set

d_set={1,2,3,4,5}
print(3 in d_set)
print(10 in d_set)
```

```python
## Set oparations
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

### Union
union_set=set1.union(set2)
print(union_set)


## Intersection
intsec_set=set1.intersection(set2)
print(intsec_set)


set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference
print(set1.difference(set2))
```

```python
## Symetric Difference 

set1.symmetric_difference(set2)
```

```python
## Set methods
set1={1,2,3}
set2={3,4,5}

## is subset
print(set1.issubset(set2))

## is superset
print(set1.issuperset(set2))
```

```python
list = [1,2,2,3,4,4,5]

# to remove all duplicates
set(list)
```

```python
### Counting Unique words in text

text="In this tutorial we are discussing about sets"
words=text.split()

## convert words to set

uni_words=set(words)
print(uni_words)
print(len(uni_words))
```
