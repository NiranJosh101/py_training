Write a function called sum_list with one parameter that takes a 
nested list of integers as an argument and returns the sum of the 
integers. For example, if you pass [[2, 4, 5, 6], [2, 3, 5, 6]] as an 
argument your function should return a sum of 33.

```python
def sum_list(nested_list):
    denested_list = []
    for value in nested_list:
        for i in value:
            denested_list.append(i)
            final_sum = sum(denested_list)
    return final_sum
```

```python
listt = [[2, 4, 5, 6], [2, 3, 5, 6], [3]]
sum_list(listt)
```

```python
def tst(nestedList):
    new_list = [x for i in nestedList for j in i for x in j]
    up_list = sum(new_list)
    return up_list
```

```python
listt = [[[2, 4, 5, 6], [2, 3, 5, 6], [3]]]
tst(listt)
```
