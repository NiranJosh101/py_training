Create a function called biggest_odd that takes a string of numbers 
and returns the biggest odd number in the list. For example, if you 
pass ‘23569’ as an argument, your function should return 9. Use list 
comprehension. 

```python
def biggest_odd(str):
    bn=[]
    for i in str:
        intt = int(i)
        if intt%2==0:
            bn.append(intt)
    return max(bn)




    

```

```python
## Using list comprehension

def biggest(strr):
    even = [x for x in strr if int(x) % 2 != 0]
    return int(max(even))
```

```python
str= '23569'
biggest_odd(str)
```

```python
strr= '23569'
biggest(strr)
```
