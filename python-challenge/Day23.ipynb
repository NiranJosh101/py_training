Create a simple calculator. The calculator should be able to perform basic 
math operations, add, subtract, divide and multiply. The calculator 
should take input from users. The calculator should be able to handle 
ZeroDivisionError, NameError, and ValueError.

```python
class Calculator:
    def __init__(self):
        pass

    def add(self, *args):
        try:
            return sum(args)
        except Exception as e:
            print(e)

    
    def subtract(self, *args):
        try:
            if not args:
                raise ValueError("At least one number is required.")
            
            sub_res = args[0]
            for num in args[1:]:
                sub_res -= num
            return sub_res
        except Exception as e:
            print(e)

    
    def multiply(self, *args):
        try:
            if not args:
                raise ValueError("At least one number is required.")
            
            mul_res = args[0]
            for num in args[1:]:
                mul_res *= num
            return mul_res
        except Exception as e:
            print(e)

    
   
    def divide(self, *args):
        if not args:
            raise ValueError("At least one number is required.")

        result = args[0]
        try:
            for num in args[1:]:
                result /= num
            return result
        except ZeroDivisionError:
            return "Error: Division by zero is not allowed."


    

    

```

```python
cal = Calculator()
```

```python
cal.divide(5, 2)
```
