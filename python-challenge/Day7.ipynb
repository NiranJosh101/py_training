Write a function called string_range that takes a single number and 
returns a string of its range. The string characters should be 
separated by dots(.) For example, if you pass 6 as an argument, 
your function should return ‘0.1.2.3.4.5’. 

```python
def string_range(num):
    rag = ''
    for i in range(num):
        rag += str(i) + '.'
    return rag[:-1]  # Remove the trailing dot
```

```python
print(string_range(6))
```

```python
def str_dot(num):
    return ".".join(str(i) for i in range(num))
```

```python
print(str_dot(6))
```

```python
def str_dash(numb):
    return "--".join(str(i) for i in range(numb))
```

```python
print(str_dash(6))
```

```python
def str_join(num):
    item = []
    for i in range(num):
        item.append(str(i))
        last = item[:-1]
    return "--".join(last)
```

```python
str_join(6)
```
