Write a function called capitalize. This function takes a string as an 
argument and capitalizes the first letter of each word. For example, 
‘i like learning’ becomes ‘I Like Learning’. 

```python
def capitalize(words):
    words.split()
    word = words[0]
    return word.capitalize() + words[1:]
```

```python
capitalize('i like learning')
```

```python
# rule based approach

def capitalize(a): 
    upper = [] 
    for i, word in enumerate(a.split()): 
        if word[0].islower(): 
            upper.append(word.capitalize()) 
        else: 
            upper.append(word) 
    return ' '.join(upper) 
print(capitalize('i love python')) 
```
