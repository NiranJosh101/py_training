```python
def average_calories(): 
    scores = [] 
    while True: 
        days = int(input("Enter the number of days you would like to estimate: ")) 
        if days <= 0: 
            print("Please enter a valid number of days.") 
            continue 
        break 
    for _ in range(days): 
        score = input('Enter a score or done when quit: ') 
        if score == 'done': 
            break 
        scores.append(int(score)) 
    return f" Mean of scores is {sum(scores) / days :.2f}" 
```
