# Day 2: Strings to Integers
Write a function called convert_add that takes a list of strings as an 
argument and converts it to integers and sums the list. For example 
[‘1’, ‘3’, ‘5’] should be converted to [1, 3, 5] and summed to 9

```python
def convert_add(list):
    integerlist=[]
    for value in list:
        integer=int(value)
        integerlist.append(integer)
        total = sum(integerlist)
    return total


## another approach
def converty_add(list):
    integerlist = [int(x) for x in list]
    total = sum(integerlist)
    return total

```

```python
# approach 1
convert_add(['1','3','5'])
```

```python
# approach 2
converty_add(['1','3','5'])
```

```python
num=["1","3","5"]

def convert_to_num(num):
    return int(num)

nun = list(map(convert_to_num,num))
total = sum(nun)

total
```
