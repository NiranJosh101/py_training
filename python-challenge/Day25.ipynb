Create a function called all_the_same that takes one argument, a 
string, a list, or a tuple and checks if all the elements are the same. 
If the elements are the same, the function should return True. If not, 
it should return False. For example, [‘Mary’, ‘Mary’, ‘Mary’] should 
return True. 

```python
from typing import Any

def all_the_same(data: Any):
    if all(x == data[0] for x in data):
        return True
    else:
        return False
```

```python
listt = ['Mary', 'Mary', 'Mary', 'Mary']

all_the_same(listt)
```

```python
from typing import Any

def all_the_same(data: Any) -> bool: # '-> bool' define the expected output
    if not data:
        return False  # Or True if you prefer empty input to be "same"
    return all(x == data[0] for x in data)
```
