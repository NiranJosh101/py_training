Write a function called equal_strings. The function takes two strings 
as arguments and compares them. If the strings are equal (if they 
have the same characters and have equal length), it should return 
True, if they are not, it should return False. For example, ‘love’ and 
‘evol’ should return True. 

```python
def equal_strings(str1, str2):

    strN = [str(i) for i in str1]
    strP = [str(e) for e in str2]


    ## use set(), sorted(), import and use counter() to check each character of the strings
    if sorted(strN) == sorted(strP) and len(strN) == len(strP):
        return True
    else:
        return False
```

```python
equal_strings('love', 'lvoe')
```

```python
def eql_str(str1, str2):
    strN = set([str(i) for i in str1])
    strP = set([str(e) for e in str2])

    if strN.intersection(strP) == strN and len(strN) == len(strP):
        return True
    else:
        return False
```

```python
eql_str('love', 'lvoe')
```
